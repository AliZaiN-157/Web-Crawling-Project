### Step 6 - Find your Data

- name
- brand 
- current price
- previous price 

```python
results_item = results_json['products']
```

```python
results_item[0]['name']
```

```python
results_item[0]['brandName']
```

```python
results_item[0]['price']['current']['text']
```

```python
results_item[0]['price']['previous']['text']
```

### Step 7 - Put Everything together
#### Loop through results and append data inside a list

```python
name = []
brand = []
current_price = []
old_price = []

for result in results_item:
    
    #name
    name.append(result['name'])
    
    #brand
    brand.append(result['brandName'])
    
    #current price
    current_price.append(result['price']['current']['text'])
    
    #old price
    old_price.append(result['price']['previous']['text'])
```

### Step 8 - Pandas DataFrame

```python
mens_fashion = pd.DataFrame({'Name':name,
                             'Brand':brand,
                             'Current Price': current_price,
                             'Old Price': old_price
                            })
```

### Step 9 - Turn DataFrame to Excel / Json

```python
mens_fashion.to_excel('mens_fashion.xlsx', index = False)
```

```python
mens_fashion.to_json('mens_fashion.json',orient="split")
```
