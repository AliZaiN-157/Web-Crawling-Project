### Step 1 - Import Packages

```python
import requests
import pandas as pd
```

### Step 2 - Requests & CURL


```python

headers = {
    'sec-ch-ua': '^\\^Chromium^\\^;v=^\\^92^\\^, ^\\^',
    'Referer': 'https://www.ebooks.com/en-pk/subjects/computers/',
    'sec-ch-ua-mobile': '?1',
    'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Mobile Safari/537.36',
    'Content-Type': 'application/json',
}

params = (
    ('subjectId', '13'),
    ('pageNumber', '1'),
    ('countryCode', 'PK'),
)

response = requests.get('https://www.ebooks.com/api/search/subject/', headers=headers, params=params)
```

### Step 6 - Find your Data

```python
# title
# subtitle
# author
# publisher
# publication date
# price
```

```python
book_results= response.json()['books']
```

```python
len(book_results)
```

### Step 7 - Put everything together
####  Loop through results and append data inside a list

```python
book_results[0]
```

```python
# title
book_results[0]['title']
# subtitle 
book_results[0]['subtitle']
# author
book_results[0]['authors'][0]['author_name']
# publisher
book_results[0]['publisher']
# publication date
book_results[0]['publication_year']
# price 
book_results[0]['price']
```

```python
title = []
subtitle = []
author = []
publisher = []
publication_year = []
price = []

for results in book_results:
    title.append(results['title'])
    subtitle.append(results['subtitle'])
    author.append(results['authors'][0]['author_name'])
    publisher.append(results['publisher'])
    publication_year.append(results['publication_year'])
    price.append(results['price'])
```

### Step 8 - Create Dataframe

```python
books_df = pd.DataFrame({'Title':title, 
                         'Subtitle':subtitle, 
                         'Author':author, 
                         'Publisher':publisher, 
                         'Publication Year':publication_year, 
                         'Price':price
                        })
```

### Step 9 - Dataframe to Excel


```python
books_df.to_excel('books.xlsx', index=False)
```
