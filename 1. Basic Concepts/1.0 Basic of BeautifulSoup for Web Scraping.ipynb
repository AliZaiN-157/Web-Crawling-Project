### Imports

```python
from bs4 import BeautifulSoup
import requests
```

```python
# docs
# https://www.crummy.com/software/BeautifulSoup/bs4/doc/#find
```

### HTTP Request

```python
# store website in a variable
website = "http://automationpractice.com/index.php?id_category=3&controller=category"
```

```python
# storing request in reponse
response = requests.get(website)
```

```python
#check status code
response.status_code
```

### Soup Object 

```python
soup = BeautifulSoup(response.content, 'html.parser')
#print(soup)
```

### Important Functions

#### find() 

return first element

```python
# example 1 - find by id
contact_us = soup.find(id='contact-link').get_text().strip()
contact_us
```

```python
# example 2 - find by id
contact_us_2 = soup.find(id='contact-link').find('a')
contact_us_2
```

```python
#example 3 - find by class
result = soup.find(class_ = 'ajax_block_product').find(class_ ='content_price')
result
```

##### find_all() & findAll()

return a list


```python
# example - by find_all()
result_ = soup.find_all(class_ = 'ajax_block_product')
result_
```

```python
len(result_)
```

```python
# example - by findAll()
result_x = soup.findAll(class_ = 'ajax_block_product')
result_x
```

```python
len(result_x)
```

#### select_one()

select by css selector - returns first element - equivalent to "find()"

```python
soup.select_one('.shop-phone').get_text().strip().replace('Call us now: ','') #to target class use .
```

```python
soup.select_one('#contact-link').get_text().strip()
```

#### select()

select by css selector - returns a list -equivalent to "find_all() / findAll()"

```python
result_y = soup.select('.ajax_block_product')
result_y
```

#### Extract Attribute Values vs get_text()

```python
#find by attribute
soup.find(itemprop='name').find('a').get('title')
```

```python
soup.find(itemprop= 'name').find('a').get_text().strip()
```

#### Siblings and Parents

```python
#go to product #1 and find the next sibling
soup.select_one('.ajax_block_product').find_next_sibling().select_one('.product-name').get_text().strip()
```

```python
#go to product #7 and find the previous sibling
soup.select('.ajax_block_product')[6].find_previous_sibling().select_one('.product-name').get_text().strip()
```

```python
#go to last product and find parent element
soup.select('.ajax_block_product')[6].find_parent()
```

#### Extract Links & URLs

Singel Link

```python
soup.find(id="contact-link").find('a')['href']
```

Multiple Link

```python
link_variable = soup.find_all(itemprop='name')
link_variable
```

```python
links = []
for a in link_variable:
    links.append(a.find('a')['href'])
```

```python
links
```
